# EDA: Gender Inequality and Maternal Mortality

One-off exploration to validate pipeline assumptions before building the dashboard.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import plotly.express as px

from src.data_cleaning import clean_and_merge, get_snapshot
from src.metrics import compute_gap_score
from src.clustering import run_clustering

pd.set_option('display.max_columns', None)
```

```python
df = clean_and_merge()
df.shape, df['Year'].min(), df['Year'].max()
```

```python
# Coverage per variable
df[['maternal_mortality_ratio', 'gii', 'gdi']].notna().sum()
```

```python
snapshot = get_snapshot(df, 'maternal_mortality_ratio', 'gii')
print('Year used as snapshot:', snapshot['Year'].iloc[0])
snapshot.head()
```

```python
fig = px.scatter(
    snapshot, x='gii', y='maternal_mortality_ratio',
    color='region', hover_name='Entity', log_y=True,
    title='GII vs. maternal mortality (snapshot)'
)
fig.show()
```

```python
scored = compute_gap_score(snapshot, outcome_col='maternal_mortality_ratio', indicator_col='gii')
scored.nlargest(10, 'gap_score')[['Entity', 'gii', 'maternal_mortality_ratio', 'gap_score']]
```

```python
clustered, meta = run_clustering(snapshot, outcome_col='maternal_mortality_ratio', indicator_col='gii')
print('k chosen:', meta['k'])
print('silhouette scores:', meta['silhouette_scores'])
clustered['cluster_label'].value_counts()
```
